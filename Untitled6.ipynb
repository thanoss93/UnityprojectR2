```python
import pandas as pd
import json
from pathlib import Path
from ast import literal_eval

# path to file
p = Path(r'//Users/stevetoner/Downloads/data_zipped.json')

list_of_dicts = list()
with p.open('r', encoding='utf-8') as f:
    data = f.readlines()
    for x in data:
        values_to_fix = {'false': 'False',
                         'true': 'True',
                         'none': 'None'}
        for k, v in values_to_fix.items():
            x = x.replace(k, v)
        try:
            x = literal_eval(x)
            list_of_dicts.append(x)
        except ValueError as e:
            print(e)
            print(x)

df = pd.json_normalize(list_of_dicts)
```
